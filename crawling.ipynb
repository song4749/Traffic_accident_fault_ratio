```python
import requests
from bs4 import BeautifulSoup
import json
import time
import urllib3
```

```python
# 데이터 저장 리스트
accident_data = []

for page_number in range(1, 62):
    sub_page_number = 1
    
    while True:
        url = f"https://accident.knia.or.kr/myaccident-content?chartNo=%EC%B0%A8{page_number}-{sub_page_number}&chartType=1"
        
        # SSL 인증서 경고 무시
        urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

        # 페이지 요청
        response = requests.get(url, verify=False)
        print(url)
        
        # 페이지가 존재하지 않으면 종료
        if "요청하신 페이지를 찾을 수 없습니다" in response.text:
            break

        # HTML 파싱
        soup = BeautifulSoup(response.text, "html.parser")

        # 특정 텍스트 추출
        car_A = soup.select_one(".cont_l .con")
        car_B = soup.select_one(".cont_r .con")
        accident_description = soup.select_one("#smrizeexplna")
        fault_A = soup.select_one("td .red")
        fault_B = soup.select_one("td .orange")

        # 텍스트에서 줄 바꿈 문자 제거
        description_text = accident_description.text.strip().replace("\r\n", " ").replace("\n", " ").replace("⊙ ", "")
    
        if car_A and car_B and accident_description and fault_A and fault_B:
            accident_data.append({
                "사고유형": f"차{page_number}-{sub_page_number}",
                "자동차 A": car_A.text.strip(),
                "자동차 B": car_B.text.strip(),
                "사고 설명": description_text,
                "과실 비율": f"{fault_A.text.strip()} : {fault_B.text.strip()}"
            })

        sub_page_number += 1
        time.sleep(1)  # 서버 부하 방지를 위해 1초 대기
```

```python
# JSON 파일 저장
with open("accident_data.json", "w", encoding="utf-8") as f:
    json.dump(accident_data, f, ensure_ascii=False, indent=4)

print("✅ 크롤링 완료! accident_data.json 파일이 생성되었습니다.")
```
